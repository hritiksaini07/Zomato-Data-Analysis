```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato-data-.csv")
print(dataframe.head())
```

```python
#to get the rate value in float removing the denominator
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()    #to obtain the summary of the dataframe
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Restaurant")

#to categorize the restaurants and find out what category majority restaurants are of
```

```python
#to check the preference of the people according to restaurant type
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'Votes':grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel("Type of Restaurant", c='red', size=20)
plt.ylabel("Votes", c='red', size=20)

```

```python
#to find out the restaurant(s) with maximum votes
max_votes=dataframe['votes'].max()
restaurant_with_max_votes= dataframe.loc[dataframe['votes']==max_votes,'name']
print("Restaurant(s) with maximum votes:", restaurant_with_max_votes)
```

```python
#to check what percentage of restaurants accept online orders
sns.countplot(x=dataframe['online_order'])
```

```python
#to check the rating distribution 
plt.hist(dataframe['rate'],bins=5)%%
plt.title('Ratings Distribution')
plt.show()
```

```python
#to find average cost of two people 
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
#rating received based on order type (online/offline)
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order', y='rate', data=dataframe)
```

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```
